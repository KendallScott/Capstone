```python
df.contact.unique()
```

```python
df['state'] = np.where(df['contact'] == 1,'VA',
                       np.where(df['contact'] == 2,'WV',
                                np.where(df['contact'] == 3,'KY',
                                         np.where(df['contact'] == 4,'TN',
                                                  np.where(df['contact'] == 8,'AL','N/A')))))
```

```python
df.state.unique()
```
